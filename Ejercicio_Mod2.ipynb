## Histograma y Curvas de Densidad

```python
def histograma(sub, var, la, mean, medi, maxi, mini, r, tit):
    plt.subplot(1, 2, sub)
    plt.hist(var, bins=50, color='blue', alpha=0.4)
    plt.xlabel("la")
    plt.ylabel("Cantidad viviendas")
    plt.axvline(mean, color='red', linestyle= 'dashed', linewidth=1, label=(f"Media: {mean:.2f} {r}"))
    plt.axvline(medi, color='green', linestyle= 'dashed', linewidth=1, label=(f"Mediana: {medi:.2f} {r}"))
    plt.axvline(maxi, color='black', linestyle= 'dashed', linewidth=1, label=(f"Maximo: {maxi:.2f} {r}"))
    plt.axvline(mini, color='orange', linestyle= 'dashed', linewidth=1, label=(f"Minimo: {mini:.2f} {r}"))
    plt.title(f"Histograma {tit}")
    plt.legend()
    plt.show
    
```

```python


plt.figure(figsize=(16, 4))
histograma(1, price, "Precio", mean_price, medi_price, max_price, min_price, "€", "Precios")
histograma(2, sqft, "Dimensiones", mean_sqft, medi_sqft, max_sqft, min_sqft, "m2", "m2")

plt.figure(figsize=(16, 4))
histograma(1, rooms, "Habitaciones", mean_rooms, medi_rooms, max_rooms, min_rooms, "", "Habitaciones")
histograma(2, baths, "Cuartos de Baño", mean_baths, medi_baths, max_baths, min_baths, "", "Cuartos de baño")


```


## Filtro de Outliers

```python

def filtrar_outliers(array, k, Q1, Q3, IQR):
    limite_inferior = Q1 - IQR * k
    limite_superior = Q3 + IQR * k
    array_sin_outliers = array[(array >= limite_inferior) & (array <= limite_superior)]
    return array_sin_outliers

```

```python
precio_sin_outliers = filtrar_outliers(price, 1.5, Q1_price, Q3_price, iqr_price)
hab_sin_outliers = filtrar_outliers(rooms, 1.5, Q1_room, Q3_room, iqr_room)
baños_sin_outliers = filtrar_outliers(baths, 1.5, Q1_bath, Q3_bath, iqr_bath)
m2_sin_outliers = filtrar_outliers(sqft, 1.5, Q1_m2, Q3_m2, iqr_m2)

```

## Correlación

```python
columnas = [price, rooms, baths, sqft]
matriz_lista = list()
for columna1 in columnas:
    fila = list()
    for columna2 in columnas:
        fila.append(stats.pearsonr(columna1, columna2)[0])  
    matriz_lista.append(fila)
    
matriz = np.array(matriz_lista)
print(matriz)
```

## Contraste de hipótesis:

Que las casas de X barrio son más baratas de las de Y barrio
Que las casas de >= 3 baños son más caras que las casas de 1-2 baños

* Contraste Hipótesis 1

```python
from scipy import stats


barrio_x = "Recoletos"
barrio_y = "Hortaleza"
precios_x = price[address == barrio_x]
precios_y = price[address == barrio_y]


stat1, pvalue1 = stats.ttest_ind(precios_x, precios_y, alternative="greater")

print("\nContraste de hipótesis 1: Precios en", barrio_x, "vs", barrio_y)
print("Estadistico:", stat1, "P-valor:", pvalue1)

if pvalue1 <0.05: 
    print("Rechazamos H0: Las casas en",barrio_x, "son más caras que en", barrio_y)
else:
    print("No podemos rechazar H0: No hay evidencia que las casas en ", barrio_x, "sean más caras", barrio_y)

```

* Contraste Hipótesis 2

```python
precios_3_baños = price[baths >3]
precios_1_2_baños = price[(baths == 1) | (baths == 2)]

stat2, pvalue2 = stats.ttest_ind(precios_3_baños, precios_1_2_baños, alternative="greater")
print("Estadistico", stat2, "P-valor:", pvalue2)

if pvalue2 <0.05:
    print("Rechazamos H0: Las casas son >=3 ba;os son más caras que las de 1-2 baños")
else:
    print("No podemos rechazar H0: No hay evidencia de que las casas con >=3 baños sean más caras")
    
```
